```python
device = "cpu"
default_tensor_type = torch.FloatTensor
def kendall_tau(y, sigma):
    """
    Kendall Tau distance
    Permutations follow the convention rank -> item_id
    :param torch.Tensor y: permutation
    :param torch.Tensor sigma: permutation
    :return: value of the metric
    :rtype: torch.Tensor
    """
    with torch.no_grad():
        n = sigma.size()[-1]
        sigma_inv = torch.argsort(sigma, dim=-1).to(device).type(default_tensor_type)
        y_inv = torch.argsort(y, dim=-1).to(device).type(default_tensor_type)
        sigma_pairs = (sigma_inv.unsqueeze(dim=-1) > sigma_inv.unsqueeze(dim=-2)).to(device).type(default_tensor_type)
        y_pairs = (y_inv.unsqueeze(dim=-1) > y_inv.unsqueeze(dim=-2)).to(device).type(default_tensor_type)
        return torch.sum(torch.abs(sigma_pairs-y_pairs), dim=[-2,-1])/(n*(n-1))
```

```python
t_ = 0
w_ = 1.0

def get_dist(t_, w_):
    keyt = "threshold_"+str(t_)
    keyw = "weight_"+str(w_)
    output[keyt][keyw]

    real_median = torch.tensor([4, 3, 2, 1, 0])

    dist = list()
    for elem in output[keyt][keyw]:
        d = kendall_tau(elem, real_median)
        dist.append(np.round(d.numpy().item(),3))
    return dist
```

```python
t = [0, 0.2, 0.5, 0.8]
w = [1.0, 0.9, 0.8]

for w_ in w:
    for t_ in t:
        dist = get_dist(t_, w_)
        
        plt.hist(dist)
        plt.title(f"Threshold {t_} / Weight {w_} -- Mean {np.round(np.mean(dist),2)}, std = {np.round(np.std(dist),2)}")
        plt.show()
```
